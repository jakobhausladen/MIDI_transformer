```python
import keras
from midi_utils.train import load_and_prepare_data, create_model, train_model, generate_examples
```

```python
training_file = 'training_data/english_pieces/6_bars_aug/training_data.json'
validation_file = 'training_data/english_pieces/6_bars_aug/validation_data.json'
tokenizer_file = 'training_data/english_pieces/6_bars_aug/tokenizer.json'
output_directory = 'models/model_v9'

# data parameters

SEQUENCE_LENGTH = 350
MIN_NUM_TOKENS = 100
BATCH_SIZE = 64


# model parameters

EMBEDDING_DIM = 84
FEED_FORWARD_DIM = 320
NUM_HEADS = 10
NUM_LAYERS = 10
DROPOUT = 0.1

EPOCHS = 40



tokenizer, training, validation = load_and_prepare_data(
    training_file,
    validation_file,
    tokenizer_file,
    SEQUENCE_LENGTH,
    MIN_NUM_TOKENS,
    BATCH_SIZE
)


'''
model = create_model(
    tokenizer,
    sequence_length,
    embedding_dim,
    feed_forward_dim,
    num_heads,
    num_layers,
    dropout
)
'''

model = keras.saving.load_model('models/model_v8/model.keras')

model.summary()


model = train_model(
    output_directory,
    model,
    training,
    validation,
    EPOCHS,
    patience=10
)


results = generate_examples(
    output_directory,
    model,
    tokenizer,
    max_len=SEQUENCE_LENGTH
)
```
